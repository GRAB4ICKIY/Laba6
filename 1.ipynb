```python
class NumberChecker:
    def __init__(self):
        self.data = pd.DataFrame(columns=['A', 'B', 'C'])

    def input_numbers(self, a, b, c):
        new_data = pd.DataFrame({'A': [a], 'B': [b], 'C': [c] })
        self.data = pd.concat([self.data, new_data], ignore_index=True)
    
    def is_one_even(self):
        return (self.data['A'] % 2 == 0) != (self.data['B'] % 2 == 0)

    def are_all_multiples_of_three(self):
        return (self.data['A'] % 3 == 0) & (self.data['B'] % 3 == 0) & (self.data['C'] % 3 == 0)

    def visualize_results(self):
        results = {
            'Только одно чётное': self.is_one_even().astype(int),
            'Все кратны 3': self.are_all_multiples_of_three().astype(int)
        }
        
        results_df = pd.DataFrame(results)

        plt.figure(figsize=(10, 5))
        sns.countplot(data=results_df.melt(), x='variable', hue='value', palette='Set1')
        plt.title('Результаты проверки условий')
        plt.ylabel('Количество')
        plt.xlabel('Условия')
        plt.xticks(rotation=45)
        plt.legend(title='Результат', loc='upper right', labels=['Ложь', 'Истина'])
        plt.show()

if __name__ == '__main__':
    checker = NumberChecker()

    A = int(input("Введите число A: "))
    B = int(input("Введите число B: "))
    C = int(input("Введите число C: "))

    checker.input_numbers(A, B, C)

    print("Только одно из чисел A и B четное:", checker.is_one_even().any())
    print("Каждое из чисел A, B, C кратно трем:", checker.are_all_multiples_of_three().any())

    checker.visualize_results()
```

```python
class NumberConditions:
    def __init__(self,start,end):
        self.numbers = pd.DataFrame({'numbers':np.arange(start, end + 1)})

    def check_conditions(self):
        self.numbers['condition_a'] = (self.numbers['numbers'] % 4 == 0) | (self.numbers['numbers'] % 7 == 0)
        self.numbers['condition_b'] = (self.numbers['numbers'] % 5 == 0) | (self.numbers['numbers'] % 10 == 0)

    def get_condition_a(self):
        return self.numbers[self.numbers['condition_a']]
    def get_condition_b(self):
        return self.numbers[self.numbers['condition_b']]
    
    def visual_conditions(self):
        plt.figure(figsize = (12,6))

        plt.subplot(1,2,1)
        sns.scatterplot(data = self.numbers, x = 'numbers', y = 'condition_a', hue='condition_a',palette={True:'green',False:'red'})
        plt.title('Условия а : Кратно 4 или 7')
        plt.xlabel('Числа')
        plt.ylabel('Условие а (истина/ложь)')

        plt.subplot(1,2,2)
        sns.scatterplot(data = self.numbers, x = 'numbers', y = 'condition_b', hue='condition_b',palette={True:'green',False:'red'})
        plt.title('Условия б : Кратно 5 и не оканчивается на 0')
        plt.xlabel('Числа')
        plt.ylabel('Условие b (истина/ложь)')

        plt.tight_layout()
        plt.show()
    
number_conditions = NumberConditions(1,100)
number_conditions.check_conditions()

print("Условие а (кратно 4 или 7)")
print(number_conditions.get_condition_a())

print("\nУсловие b (кратно 5 и не оканчивается на 0)")
print(number_conditions.get_condition_b())

number_conditions.visual_conditions()
```

```python
class NumberCondition:
    def __init__(self, n, m, k, l):
        self.n = n
        self.m = m
        self.k = k
        self.l = l

    def check_conditions(self):
        condition1 = self.n > 1
        condition2 = (self.m <= self.l + self.k == 0)
        condition3 = (self.n > 2 and self.m ** 2 > self.l ** 2)
        
        return [condition1, condition2, condition3]

n = int(input('Введите значение n: '))
m = int(input('Введите значение m: '))
k = int(input('Введите значение k: '))
l = int(input('Введите значение l: '))

number_condition = NumberCondition(n, m, k, l)
results = number_condition.check_conditions()

conditions_labels = ["n > 1", "m <= l + k == 0", "m^2 > l^2 (если n > 2)"]
df = pd.DataFrame({"Условие": conditions_labels, "Результат": results})

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="Условие", hue="Результат", palette='Set1')
plt.title("Проверка условий")
plt.xlabel("Условие")
plt.ylabel("Количество")
plt.show()
```

```python
class MyClassTask:
    def __init__(self, deposit):
        self.deposit = deposit

    def calculate_payout(self):
        if self.deposit < 50000:
            rate = 0.16
        elif self.deposit <= 100000:
            rate = 0.18
        else:
            rate = 0  
        
        return self.deposit * (1 + rate)

deposit = float(input('Введите сумму вклада: '))

my_task = MyClassTask(deposit)
payout = my_task.calculate_payout()

print(f"Сумма выплаты по депозиту: {payout:.2f} руб.")
```
